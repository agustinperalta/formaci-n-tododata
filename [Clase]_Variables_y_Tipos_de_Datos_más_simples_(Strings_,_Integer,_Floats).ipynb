En Python se pueden asignar valores a más de una variable en una sola línea de código. Siempre y cuando el número de valores a la derecha (N) sea igual al numero de variables a la izquiera (n).

SIEMPRE n=N

```python
a, b, c = 10, 5, 2
print("La variable a tiene valor",a)
print("La variable b tiene valor",b)
print(f"La variable c tiene valor {c}")
```
